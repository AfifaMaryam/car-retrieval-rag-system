CREATE YOLO FOLDERS

```python
import os

base = "/kaggle/working/mydata_yolo"

paths = [
    f"{base}/images/train",
    f"{base}/images/val",
    f"{base}/labels/train",
    f"{base}/labels/val"
]

for p in paths:
    os.makedirs(p, exist_ok=True)

print("Folders created")
```

Copy Images & Labels from Input

```python
import os
import shutil
import random

src_img = "/kaggle/input/yolo-v5-car-object-detection/mydata/images"
src_lbl = "/kaggle/input/yolo-v5-car-object-detection/mydata/labels"

base = "/kaggle/working/mydata_yolo"

os.makedirs(f"{base}/images/train", exist_ok=True)
os.makedirs(f"{base}/images/val", exist_ok=True)
os.makedirs(f"{base}/labels/train", exist_ok=True)
os.makedirs(f"{base}/labels/val", exist_ok=True)

images = [f for f in os.listdir(src_img) if f.endswith(".jpg")]
random.shuffle(images)

split_idx = int(0.8 * len(images))
train_imgs = images[:split_idx]
val_imgs = images[split_idx:]

for img in train_imgs:
    shutil.copy(f"{src_img}/{img}", f"{base}/images/train/{img}")
    shutil.copy(
        f"{src_lbl}/{img.replace('.jpg', '.txt')}",
        f"{base}/labels/train/{img.replace('.jpg', '.txt')}"
    )

for img in val_imgs:
    shutil.copy(f"{src_img}/{img}", f"{base}/images/val/{img}")
    shutil.copy(
        f"{src_lbl}/{img.replace('.jpg', '.txt')}",
        f"{base}/labels/val/{img.replace('.jpg', '.txt')}"
    )

print(" Train images:", len(train_imgs))
print(" Val images:", len(val_imgs))
```

Verify Dataset Integrity

```python
print("Train images:", len(os.listdir(f"{base}/images/train")))
print("Train labels:", len(os.listdir(f"{base}/labels/train")))
print("Val images:", len(os.listdir(f"{base}/images/val")))
print("Val labels:", len(os.listdir(f"{base}/labels/val")))
```

Create Dataset YAML

```python
yaml_text = f"""
path: {base}
train: images/train
val: images/val

nc: 1
names: ['car']
"""

with open("/kaggle/working/car_dataset.yaml", "w") as f:
    f.write(yaml_text.strip())

print("car_dataset.yaml created")
```
